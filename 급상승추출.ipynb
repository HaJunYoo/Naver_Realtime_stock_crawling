```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import threading
import time
from datetime import datetime, timedelta
import xlsxwriter
import re


driver = webdriver.Chrome()
# 객체 생성 


driver.implicitly_wait(2)
#driver.get('https://datalab.naver.com/keyword/realtimeList.naver?where=main')

hour = []
minute = []

for i in range(1): 
    for j in range(9,10) :
        hour.append('{}{}'.format(i,j))
        
for l in range(10,13) :
    hour.append('{}'.format(l))
        

for x in range(6) : 
    for k in range(10) :
        minute.append('{}{}'.format(x,k))
        
urllist = []

for x in hour:
    for y in minute:
        urllist.append("https://datalab.naver.com/keyword/realtimeList.naver?age=all&datetime=2020-10-15T{}%3A{}%3A00".format(x,y))
        ## 2020-월-일T시간%A{분}%3A00
    
    

xpath = [
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/div/div/ul/li[1]/div/div/a[3]',#이슈별 묶어보기
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/div/div/ul/li[2]/div/div/a[3]',#이벤트 할인
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/div/div/ul/li[3]/div/div/a[3]',#시사
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/div/div/ul/li[4]/div/div/a[3]',#엔터
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[2]/div[1]/div[1]/div/div/ul/li[5]/div/div/a[3]',#스포츠
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[1]/div[1]/div/div/div/div[1]/a[3]/span[1]', #날짜
     '/html/body/div[1]/div[2]/div[1]/div/div[2]/div[1]/div[2]/div/div/div/ul/li[6]/a',#전연령
]


        
all_word = []

    
for i in urllist:
      
    
    driver.get(i)
    
    time.sleep(1)

    for p in xpath:
        button = driver.find_element_by_xpath(p)
        driver.execute_script("arguments[0].click();", button)
   

    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    age_data = soup.find_all("span", class_ = "item_title")
    
    pop_10 = []
    
    for i in age_data :
        text = i.text
        pop_10.append(text)
        
    all_word.append(pop_10)
    
    print(all_word)
        

all_word
    
    
    
time.sleep(1)   





#driver.get() 
##여기서 시간을 변경해야하는데....
# 변수를 만들어서 시간을 삽입해야할 것 같음
# 반복문을 만들어서 그 게 들어갈 수 있게 






# 열(시간) 별로 데이터 프레임을 만들어서 csv를 만들어냄 
```
